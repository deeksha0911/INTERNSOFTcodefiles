```
class Person:
  def __init__(self,name,surname,yob):
    self.name=name
    self.surname=surname
    self.yob=yob
  def age(self,current_yr):
    return current_yr-self.yob  
  def __str__(self):
    return "%s %s was born in %d"%(self.name,self.surname,self.yob) 
alec=Person("Alec","Baldwin",1990)
print(alec)
print(alec.age(2020))    
```

```
class Student(Person):
  def __init__(self,std_id,*args,**kwargs):
    super(Student,self).__init__(*args,**kwargs)
    self._std_id=std_id
Deepthi=Student(1,'Deepthi','S',2003)
print(Deepthi) 
print(type(Deepthi))
print(isinstance(Deepthi,Person))
print(isinstance(Deepthi,object))
```
