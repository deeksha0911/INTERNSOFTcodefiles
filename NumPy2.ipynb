```
import time
import numpy as np
M=32
N=64
x=np.linspace(0,1,M,endpoint=False)
y=np.linspace(0,1,N)
x
```

```
y
```

```
t0=time.time()
g=np.zeros((M,N))
for i in range(M):
  for j in range(N):
    g[i,j]=np.sin(2*np.pi*x[i]*y[j])
t1=time.time()
print("time elapsed:{} s".format(t1-t0))    
    
```
