```
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
plt.plot(np.random.normal(size=130), np.random.normal(size=130),'d')
```

```
plt.plot(np.random.normal(size=130), np.random.normal(size=130),'o')
```

```
nor = pd.Series(np.random.normal(size=30))
print(nor)
nor.plot()
```

```
normals = pd.Series(np.random.normal(size=30))
print(normals)
normals.plot()
```

```
nor.cumsum().plot(grid=True)
```

```
a=pd.DataFrame({"normal":np.random.normal(size=20),
               'gamma':np.random.normal(1,size=20),
                "poission":np.random.normal(size=20)})
print(a)
a.cumsum(0).plot() 
```

```
a.cumsum(0).plot(subplots=True)
```

```
a.cumsum(0).plot(secondary_y='normal')
```

```
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(4, 12))
for i,var in enumerate(['normal','gamma','poission']):
  [var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
    
```

```
fig.savefig('figure.png')
```

```
fig = plt.figure()
ax = plt.axes()

s = np.linspace(0, 20, 100)
print(s)
ax.plot(s, np.sin(s));
```

```
plt.plot(s, np.sin(s - 0), color='red')        
plt.plot(s, np.sin(s - 1), color='0.52')           
plt.plot(s, np.sin(s - 2), color='0.55')       
```

```
plt.plot(s, np.sin(s - 1), color='red')        
plt.plot(s, np.sin(s - 1), color='blue')           
plt.plot(s, np.sin(s - 2), color='0.55')       
```

```
plt.plot(s, np.sin(s - 0), linestyle='-')        
plt.plot(s, np.sin(s - 1), linestyle='')           
plt.plot(s, np.sin(s - 6), linestyle='-.')  
```

```
plt.plot(s, np.sin(s - 0))

plt.xlim(-1, 15)
plt.ylim(-2, 2);
```

```
plt.plot(s, np.sin(s))
plt.title("Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```
