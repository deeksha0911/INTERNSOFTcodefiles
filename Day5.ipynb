```
try:
    a=open("filetest",'w')
    a.write('hello')
except:
  print("ERROR!!!")
else:
  print("content written")
a.close()
```

```
try:
    a=open("filetest",'r')
    a.write('hello')
except:
  print("ERROR!!!")
else:
  print("content written")
a.close()
```

```
try:
    a=open("filetest",'w')
    a.write('hello')
finally:
  print("!!")
```
